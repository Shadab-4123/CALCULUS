```python
import math
import matplotlib.pyplot as plt
```

```python
n = int(input('enter number of observation you want: '))   # number of observation you want to graph
x0 = float(input('enter x at t=0: '))      # enter initial position at t=0
v0 = float(input('enetr v at t=0: '))      # enter initial velocity at t=0
interval = float(input('enter interval: ')) # there should be proper coordination between n and interval to get good range of graph
     
```

```python
def Euler(n, interval, x, v):
  # n = 501        # x_0 = 1      # interval = 0.01 
  xlist1 = []
  for _ in range(n):
    xlist1.append(x)
    v = x*interval+v
    x = v*interval + x             # formula of Euler's method 
  return xlist1
     
```

```python
def Exponent(n):
  xlist2 = []
  for t in range(n):
    xlist2.append(math.exp(t*interval))
  return xlist2
```

```python
tlist = [i*interval for i in range(n)]
plt.plot(tlist,Euler(n,interval,x0,v0))
plt.plot(tlist,Exponent(n))
```

```python
# Error camparison between Euler and exponent graph
def Euler_error(n,interval,x,v):
  # n = 101     # interval = 0.01   # x = 1    # v = 1
  delta_t = [i*interval for i in range(n)]
  xlist1 = []          # numerical value
  xlist2 = []          # analytical value
  for t in delta_t:
    v = x*interval+v
    x = v*interval + x
    xlist1.append(x)
    xlist2.append(math.exp(t))
  error_list1 = []
  for i in range(n):
    error = (xlist1[i]-xlist2[i])/xlist1[i]
    error_list1.append(error)
  return error_list1
```

```python
delta_t = [i*interval for i in range(n)]
plt.plot(delta_t, Euler_error(n,interval,x0,v0))
```

```python
def Euler_mid(n,interval,x,v):
  # n = 501        # x_0 = 1      # interval = 0.01  # v = 1    # for good graph
  xlist1 = []
  for _ in range(n):
    xlist1.append(x)
    v = v + interval*(x+interval*x/2)
    x = x + interval*(v+interval*v/2)             # using the formula of Euler's method to find next term
  return xlist1
```

```python

tlist = [i*interval for i in range(n)]
plt.plot(tlist,Euler_mid(n,interval,x0,v0))
plt.plot(tlist,Exponent(n))
```

```python

# Error camparison between Euler's mid-point and exponent graph
n = 101      # number of observation you want to graph
def Euler_mid_error(n,interval,x,v):
  # n = 101     # interval = 0.01   # x = 1    # v = 1
  delta_t = [i*interval for i in range(n)]
  xlist1 = []          # numerical value
  xlist2 = []          # analytical value
  for t in delta_t:
    xlist1.append(x)
    v = v + t*(x+t*x/2)
    x = x + t*(v+t*v/2)
    xlist2.append(math.exp(t))
  error_list2 = []
  for i in range(n):
    error = (xlist1[i]-xlist2[i])/xlist1[i]
    error_list2.append(error)
  return error_list2
```

```python
tlist = [i*interval for i in range(101)]
plt.plot(tlist,Euler_mid_error(101,interval,x0,v0))
```
